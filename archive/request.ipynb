```python
data = request.big_data
```

```python
data["day_name"] = data["creation_date_order"].apply(lambda x : pd.Timestamp(x).day_name())
```

```python
data["day_of_month"] = data["creation_date_order"].apply(lambda x : pd.Timestamp(x).day_of_month)
```

```python
data["day_of_month"] = data["creation_date_order"].apply(lambda x : pd.Timestamp(x).day_of_week)
```

```python
data["day_of_year"] = data["creation_date_order"].apply(lambda x : pd.Timestamp(x).day_of_year)
```

```python
data.loc[:,"month_name"] = data.loc[:,"creation_date_order"].apply(lambda x : pd.Timestamp(x).month_name())
```

```python
data.loc[:,"month"] = data.loc[:,"creation_date_order"].apply(lambda x : pd.Timestamp(x).month)
```

```python
myrestaurant = data[data["year"] == 2018]
myrestaurant = myrestaurant[myrestaurant["month_name"] == "January"]
myrestaurant[["restaurant_name","real_price"]].groupby("restaurant_name").sum().sort_values("real_price",ascending=False)
```

```python
myrestaurant = data[data["year"] == 2018]
myrestaurant = myrestaurant[myrestaurant["month_name"] == "January"]
myrestaurant = myrestaurant[myrestaurant["restaurant_name"] == "Agora"]
result = myrestaurant[["restaurant_name","real_price"]].groupby("restaurant_name").sum().sort_values("real_price",ascending=False)
result.loc["Agora","real_price"]
```
