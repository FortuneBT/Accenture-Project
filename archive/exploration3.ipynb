```python
my_year = 2017
restaurant_name = "S.S. Constitution"

big_data = request.big_data

begin = str(my_year) + "-01-01"

end = str(my_year) + "-12-31"

mask = (big_data["creation_date_order"] >=  begin) & (big_data["creation_date_order"] <= end)
year = big_data.loc[mask]
my_restaurant = year[year["restaurant_name"] == restaurant_name]
newData = my_restaurant.loc[:,["restaurant_name","real_price"]].groupby("restaurant_name").sum()

newData
```
